```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = tf.keras.datasets.mnist
```

```python
# Loading MNIST Dataset
(x_train, y_train), (x_test, y_test) = data.load_data()
```

```python
f"Dimensions of Train data: {x_train.shape}, and Test data: {x_test.shape}"
```

```python
# Normalize data to improve accuracy because NN trains well on normalized data.
x_train = x_train/255 
```

```python
# Visualizing data

fig, axis = plt.subplots(4,4, figsize=(4,4))
for i in range(4):
    for j in range(4):
        axis[i,j].imshow(x_train[np.random.randint(60000)], cmap="gray")
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=10, activation='softmax')
])
```

```python
model.compile(loss="SparseCategoricalCrossentropy", optimizer="Adam", metrics=(["accuracy"]))
```

```python
model.fit(x_train,y_train,epochs=10)
```

```python
model.evaluate(x_test, y_test)
```

```python
# Accuracy of the model on test data is 97.8%
```
